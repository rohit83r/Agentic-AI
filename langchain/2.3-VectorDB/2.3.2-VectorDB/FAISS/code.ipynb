| Metric            | Similarity Score Range | Behavior                              |
| ----------------- | ---------------------- | ------------------------------------- |
| Cosine Similarity | \[-1, 1]               | Focuses on angle only |
| L2 Distance       | \[0, )                | Focuses on **magnitude + direction**  |

| Feature               | `Flat`                | `IVF` (Inverted File Index)        | `HNSW` (Graph-based Index)          |
| --------------------- | --------------------- | ---------------------------------- | ----------------------------------- |
| Type of Search     | Exact                 | Approximate (cluster-based)        | Approximate (graph-based traversal) |
| Speed               | Slow (linear scan)    | Fast (search only in top clusters) | Very Fast (graph walk)              |


| Dataset Size              | Recommended Index                 |
| ------------------------- | --------------------------------- |
| UPTO 1L                     | `IndexFlatL2` or `IndexFlatIP`    |
| UPTO 1M                  | `IndexIVFFlat` or `IndexHNSWFlat` |
| > 1M                      | `IndexIVFPQ` or `IndexHNSWFlat`   |

```python
# RAG pipeline
# RAG pipeline = Retrieval Augmented Generation pipeline
# 3 major components of RAG pipeline
# 1. Retriever   2. Generator   3. Reader
```

```python
from langchain_community.document_loaders import PyPDFLoader
FILE_PATH="../../../2.1-DataIngestion/stroy.pdf"
loader=PyPDFLoader(FILE_PATH)
pages=loader.load()
print(pages)
```

```python
pages = []
async for page in loader.alazy_load():
    pages.append(page)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
texts = text_splitter.split_documents(pages)
```

```python
len(texts)
```

```python
index=faiss.IndexFlatL2(384)
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)
vector_store.add_documents(texts)
```

```python
retriever=vector_store.as_retriever(search_kwargs={"k": 5})
retriever.invoke("Who was Gledinning?")
```

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
from langchain_core.runnables import RunnableMap

rag_chain = (
    RunnableMap({"context": retriever | format_docs, "question": RunnablePassthrough()})
    | prompt
    | model
    | StrOutputParser()
)
```

```python
rag_chain.invoke("Who was William Wilson?")
```
